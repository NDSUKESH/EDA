```python
## top platforms
platform_counts = df['Platform'].value_counts()

total_platforms = len(df['Platform'].dropna())

platform_percentages = (platform_counts / total_platforms) * 100
platform_percentages
```

```python
## top game genres
most_common_genres = df['Genre'].value_counts().head()

genre_sales = df.groupby('Genre')['Global'].sum()


top_genres = most_common_genres.index
top_genre_sales = genre_sales.loc[top_genres]


plt.figure(figsize=(8, 8))
plt.pie(top_genre_sales, labels=top_genres, autopct='%1.1f%%', startangle=140)
plt.title('Top  Game Genres by Global Sales')
plt.show()
```

```python
df['Genre'].unique()
```

```python
sorted_df=df.sort_values(by='Year', ascending=True)
sorted_df.head(3)
```

```python
## yearly sales of each game
yearly_sales = df.groupby('Year')['Global'].sum()

plt.figure(figsize=(12, 6))
plt.plot(yearly_sales.index, yearly_sales.values, marker='o', linestyle='-', color='b')
plt.xlabel('Year')
plt.ylabel('Global Sales (in millions)')
plt.title('Trend of Global Game Sales (1980-2020)')
plt.grid(True)
plt.xticks(rotation=45)
plt.show()
```

```python
df.head(3)
```

```python
## average sales of different region
average_sales = df[['North America', 'Europe', 'Japan', 'Rest of World']].mean()

regions = average_sales.index

plt.figure(figsize=(8, 6))
plt.bar(regions, average_sales, color=['b', 'g', 'r', 'c'])
plt.xlabel('Region')
plt.ylabel('Average Sales (in millions)')
plt.title('Average Sales Across Different Regions')
plt.xticks(rotation=45)
plt.show()
```

```python
df.head(2)
```

```python
## Top 3 game publishers
publisher_sales = df.groupby('Publisher')['Global'].sum()

top_publishers = publisher_sales.nlargest(3)

plt.figure(figsize=(8, 6))
top_publishers.plot(kind='bar', color='skyblue')
plt.xlabel('Publisher')
plt.ylabel('Global Sales (in millions)')
plt.title('Top 3 Publishers by Global Sales')
plt.xticks(rotation=45)
plt.show()
```
